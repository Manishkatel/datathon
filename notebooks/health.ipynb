```python
ms_data = ms_data.drop(columns = ['Short_Question_Text']) 
```

```python
ms_data = ms_data.rename(columns={
    "CountyFIPS": "GEOID"
})  
```

```python
import geopandas as gpd

counties = gpd.read_file("https://www2.census.gov/geo/tiger/GENZ2023/shp/cb_2023_us_county_500k.zip") 
```

```python
ms_counties = counties[counties["STATEFP"] == "28"] 
```

```python
ms_counties["GEOID"] = ms_counties["GEOID"].astype(str)
ms_data["GEOID"] = ms_data["GEOID"].astype(str) 
```

```python
merged = ms_counties.merge(ms_data, on="GEOID") 
```

```python
print(ms_counties["GEOID"].dtype)
print(ms_data["GEOID"].dtype) 
```

```python
import plotly.express as px
import geopandas as gpd
import json 

geojson = json.loads(ms_counties.to_json())  
```

```python
fig = px.choropleth(
    ms_data,
    geojson=geojson,
    locations="GEOID",
    featureidkey="properties.GEOID",
    color="Data_Value",
    color_continuous_scale="YlOrRd",
    scope="usa",
    hover_name="CountyName",
    labels={"Data_Value": "Health Value"},
    title="Mississippi County Health Data"
)

fig.update_geos(fitbounds="locations", visible=False)

fig.show(renderer="browser") 
```

```python
import seaborn as sns

avg_measure = (
    ms_data
    .groupby('MeasureId')['Data_Value']
    .mean()
    .sort_values(ascending=False)
    .head(15)
)

plt.figure(figsize=(14,6))

sns.barplot(
    x=avg_measure.index,
    y=avg_measure.values
)

plt.xticks(rotation=45)

plt.title('Average Health Metric by Measure')
plt.xlabel('Measure')
plt.ylabel('Average Data Value')

plt.show() 
```

```python
pivot_table = ms_data.pivot_table(
    values='Data_Value',
    index='CountyName',
    columns='MeasureId',
    aggfunc='mean'
)

plt.figure(figsize=(18,12))

sns.heatmap(pivot_table, cmap='YlOrRd')

plt.title('County vs Health Measure Heatmap')

plt.show() 
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=ms_data,
    x='TotalPopulation',
    y='Data_Value',
    hue='Category'
)

plt.title('Population vs Health Metric')
plt.xlabel('Population')
plt.ylabel('Health Value')

plt.show() 
```

```python
county_avg = (
    ms_data
    .groupby('CountyName')['Data_Value']
    .mean()
    .reset_index()
    .sort_values(by='Data_Value', ascending=False)
)

fig = px.bar(
    county_avg.head(20),
    x='CountyName',
    y='Data_Value',
    color='Data_Value',
    color_continuous_scale='YlOrRd',
    title='Top Counties by Health Metric'
)

fig.show(renderer='browser') 
```

```python
import plotly.express as px
import plotly.io as pio

# Optional: use latest year only
latest_year = merged["Year"].max()
df = merged[merged["Year"] == latest_year].copy()

# Make sure GEOID is string
df["GEOID"] = df["GEOID"].astype(str)

# If multiple rows per county/measure, average them
map_data = (
    df.groupby(["MeasureId", "GEOID", "CountyName"], as_index=False)
      ["Data_Value"]
      .mean()
)

html_parts = []

for measure in map_data["MeasureId"].dropna().unique():

    disease_df = map_data[map_data["MeasureId"] == measure]

    fig = px.choropleth(
        disease_df,
        geojson=geojson,
        locations="GEOID",
        featureidkey="properties.GEOID",
        color="Data_Value",
        color_continuous_scale="YlOrRd",
        hover_name="CountyName",
        title=f"Mississippi County Map: {measure}",
        labels={"Data_Value": "Health Value"}
    )

    fig.update_geos(
        fitbounds="locations",
        visible=False
    )

    fig.update_layout(
        height=650,
        margin=dict(l=20, r=20, t=60, b=20)
    )

    html_parts.append(
        pio.to_html(
            fig,
            full_html=False,
            include_plotlyjs="cdn"
        )
    )

full_html = f"""
<html>
<head>
    <title>Mississippi Disease Comparison Maps</title>
</head>
<body style="font-family: Arial; background-color: #f7f7f7;">

<h1 style="text-align:center;">
    Mississippi Disease Comparison Maps — {latest_year}
</h1>

<p style="text-align:center;">
    Each map shows county-level Data_Value for one health measure.
</p>

{''.join(html_parts)}

</body>
</html>
"""

with open("mississippi_disease_comparison_maps.html", "w", encoding="utf-8") as f:
    f.write(full_html) 
```
